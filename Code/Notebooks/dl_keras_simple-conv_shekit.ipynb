```python
def imcrop_tosquare(img):
    #make any image a square
    
    if img.shape[0] > img.shape[1]: #if height more than width, reduce height
        extra = (img.shape[0] - img.shape[1])
        if extra % 2 == 0:
            crop = img[extra // 2: -extra//2, :]
        else:
            crop = img[max(0, extra //2 +1):min(-1,-(extra//2)),:]
            
    elif img.shape[1] > img.shape[0]: # if width more reduce width
        extra = (img.shape[1]- img.shape[0])
        if extra % 2 == 0:
            crop = img[:, extra//2:-extra//2]
        else:
            crop = img[:, max(0, extra//2+1):min(-1, -(extra//2))]
    else:
        crop = img
    
    return crop
```

```python
def imcrop(img, amt):
    if amt<=0 or amt>=1:
        return img
    row_i = int(img.shape[0]*amt)//2
    col_i = int(img.shape[1]*amt)//2
    
    return img[row_i:-row_i, col_i:-col_i]
```

```python
imgs = []

# makes images square, then crop away some background, then resize to final small size
for file_i in files:
    img = plt.imread(file_i)
    square = imcrop_tosquare(img)
    crop = imcrop(square, 0.2)
    rsz = imresize(crop, (64,64))
    imgs.append(rsz)
print(len(imgs))
```
