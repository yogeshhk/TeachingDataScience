```python
import json
from pathlib import Path

import pandas as pd
from llama_cpp import Llama
from sklearn.metrics import accuracy_score, classification_report
from tqdm import tqdm
```

```python
llm = Llama(
    model_path="./llama-3.2-1b-mental-health-sentiment-q8_0.gguf",
    n_gpu_layers=-1,  # all layers on the GPU
    n_ctx=4096,
    verbose=False,
)
```

```python
!gdown 1roGWeGaTIsFr_yHVkSrD5kL2dJtD7Bsq
```

```python
test_data = json.loads(Path("test_data.json").read_text())
```

```python
print(test_data[0]["input"])
print(test_data[0]["output"])
```

```python
%%time
example = test_data[0]
messages = [
    {"role": "user", "content": example["input"]},
]
response = llm.create_chat_completion(messages, max_tokens=16, temperature=0.0)
print(response["choices"][0]["message"]["content"])
```

```python
predictions = []
true_values = []
for example in tqdm(test_data):
    messages = [
        {"role": "user", "content": example["input"]},
    ]
    response = llm.create_chat_completion(messages, max_tokens=16, temperature=0.0)
    predictions.append(response["choices"][0]["message"]["content"])
    true_values.append(example["output"])
```

```python
eval_df = pd.DataFrame.from_dict({"label": true_values, "prediction": predictions})
eval_df.head()
```

```python
accuracy_score(true_values, predictions)
```

```python
print(classification_report(eval_df.label, eval_df.prediction))
```
